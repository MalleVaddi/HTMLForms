```python
import numpy as np

data2['log_income'] = np.log(data2['MonthlyIncome'])
data2
```

In the employee_rating dataset, we can convert columns `EnvironmentSatisfaction` and `PerformanceRating` to numbers. In the below command, I have mapped the values of these two columns to numbers and assigned the resulted values to the new columns `EnvironmentSatisfaction_level`, `Performance_score` respectively.

```python
EnvironmentSatisfaction_map = {
    'Low':1,
    'Medium':2,
    'High':3,
    'Very High' : 4
}
PerformanceRating_map = {
    'Excellent':1,
    'Outstanding':2
}

data2['EnvironmentSatisfaction_level'] = data2['EnvironmentSatisfaction'].map(EnvironmentSatisfaction_map)
data2['Performance_score'] = data2['PerformanceRating'].map(PerformanceRating_map)
data2
```

In the employee_general dataset, we can categorize values of columns `Education` and `MaritalStatus`. In the below command, I have mapped the values of these two columns to numbers and assigned the resulted values to the new columns Education_Category, MaritalStatus_score respectively.

```python
Education_map = {
    'Below College':1,
    'College':2,
    'Bachelor':3,
    'Master' : 4,
    'Doctor' : 5,
}
MaritalStatus_map = {
    'Single':1,
    'Married':2,
    'Divorced':3
}

data1['Education_Category'] = data1['Education'].map(Education_map)
data1['MaritalStatus_score'] = data1['MaritalStatus'].map(MaritalStatus_map)
data1
```

Merging both `employee_rating` and `employee_general `

```python
data1.merge(data2, on='EmployeeNumber')
```
